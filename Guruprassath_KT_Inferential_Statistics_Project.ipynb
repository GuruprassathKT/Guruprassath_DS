Problem 3

Zingaro stone printing is a company that specializes in printing images or patterns on polished or unpolished stones. However, for the optimum level of printing of the image, the stone surface has to have a Brinell's hardness index of at least 150. Recently, Zingaro has received a batch of polished and unpolished stones from its clients. Use the data provided to answer the following (assuming a 5% significance level);

3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

```python
df = pd.read_csv('Zingaro_Company.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

Here we can proceed with one sample t test as the sigma of the unpolished tiles is not known.

Step 1 : Define null and alternate Hypothesis

Null hypothesis states here in this case that unpolished tiles stone surface has to have a Brinell's hardness index of at least mu = 150

Alternate Hypothesis says that unpolished tiles dont have the hardness index 150

Null hypothesis H0 = unpolished tiles = 150

Alternate hypothesis Ha = unpolished tiles is not equal to 150

* $H_0$: $\mu$ = 150
* $H_A$: $\mu$ $\neq$ 150

Step 2 : Decide the significance level

Here we select $\alpha$ = 0.05.

```python
print("The sample size for this problem is",len(df))
```

Step 3 : Indentify the test statistic

We do not know the population standard deviation and n = 75, so we are going with one sample t-test as for the above question we are only considering only unpolished tiles.

Step 4 : calculate the p_value and t- statistic

```python
from scipy.stats import   ttest_1samp
t_statistic, p_value = ttest_1samp(df['Unpolished '], 150)
print('One sample t test \nt statistic: {0} p value: {1} '.format(t_statistic, p_value))
```

Step 5 : Decide to reject or accept the null hypothesis

```python
alpha_value = 0.05 
print('Level of significance: %.2f' %alpha_value)
if p_value < alpha_value: 
    print('We have evidence to reject the null hypothesis since p value < Level of significance')
else:
    print('We have no evidence to reject the null hypothesis since p value > Level of significance') 

print ("Our one-sample t-test p-value=", p_value)
```

From this above p_value we can consider to reject the null hypothesis as unpolished tiles are not having the hardness index of 150. so unpolished tiles are not suitable for printing.

3.2 Is the mean hardness of the polished and unpolished stones the same?

Here we can proceed with two sample t test because of 2 independent samples.

Step 1: Determine Null hypothesis and alternate hypothesis

Null hypothesis states mean of unpolished and polished stones are same

Alternate Hypothesis says mean of unpolished and polished stones are not same

Null hypothesis H0 = mean of unpolished tiles = mean of polished tiles
Alternate hypothesis Ha = mean of unpolished tiles is not equal to mean of polished tiles

* $H_0$: $\mu1$ = $\mu2$
* $H_A$: $\mu1$ not equal $\mu2$

Step 2 : Decide the significance level

Here we select $\alpha$ = 0.05.

```python
print("The sample size for this problem is",len(df))
```

Step 3 : Indentify the test statistic

We do not know the population standard deviation and n = 75 and as there are 2 independent samples so we are going to 2 sample t test.

Step 4 : calculate the p_value and t- statistic

```python
from scipy.stats import  ttest_ind
t_statistic, p_value = ttest_ind(df['Unpolished '],df['Treated and Polished'])
print('One sample t test \nt statistic: {0} p value: {1} '.format(t_statistic, p_value))
```

Step 5 : Decide to reject or accept the null hypothesis

```python
alpha_value = 0.05 
print('Level of significance: %.2f' %alpha_value)
if p_value < alpha_value: 
    print('We have evidence to reject the null hypothesis since p value < Level of significance')
else:
    print('We have no evidence to reject the null hypothesis since p value > Level of significance') 

print ("Our one-sample t-test p-value=", p_value)
```

from the above p_value, we can clearly see that the p_value is lesser than the alpha 0.05, so we can reject the null hypothesis saying that the mean of unpolished stones are not the same as polished stones.

Assumptions of Two way ANOVA
